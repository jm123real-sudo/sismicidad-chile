## Medición de eficiencia algorítmica

Se comparan dos implementaciones para identificar sismos con magnitud mayor o igual a 6.0:

- Implementación con recorrido fila por fila.
- Implementación vectorizada con pandas.

El objetivo es medir el tiempo de ejecución y justificar la elección del método más eficiente.

```python
def buscar_sismos_bucle(df, magnitud_minima):
    """
    Busca sismos fuertes recorriendo el DataFrame fila por fila.
    Esta implementación es menos eficiente porque evalúa cada registro individualmente en Python.
    """
    resultados = []

    for _, fila in df.iterrows():
        if fila["Magnitude"] >= magnitud_minima:
            resultados.append(fila)

    return resultados


def buscar_sismos_vectorizado(df, magnitud_minima):
    """
    Busca sismos fuertes usando operaciones vectorizadas de pandas.
    Esta implementación aprovecha optimizaciones internas y suele ser más eficiente.
    """
    return df[df["Magnitude"] >= magnitud_minima]
```

```python
tiempo_bucle = timeit.timeit(
    lambda: buscar_sismos_bucle(df, 6.0),
    number=10
)

tiempo_vectorizado = timeit.timeit(
    lambda: buscar_sismos_vectorizado(df, 6.0),
    number=10
)

print("Tiempo con bucle:", tiempo_bucle)
print("Tiempo vectorizado:", tiempo_vectorizado)
print("Diferencia:", tiempo_bucle / tiempo_vectorizado)
```

## Algoritmo recursivo: Merge Sort

Para demostrar la aplicación de recursividad se implementa el algoritmo Merge Sort sobre la variable Magnitude.

Este algoritmo utiliza la estrategia divide y vencerás, dividiendo el problema en subproblemas más pequeños hasta alcanzar un caso base y posteriormente combinando los resultados ordenados.

```python
def merge_sort(lista):

    if len(lista) <= 1:
        return lista

    medio = len(lista) // 2

    izquierda = merge_sort(lista[:medio])
    derecha = merge_sort(lista[medio:])

    return combinar(izquierda, derecha)


def combinar(izquierda, derecha):

    resultado = []
    i = 0
    j = 0

    while i < len(izquierda) and j < len(derecha):

        if izquierda[i] <= derecha[j]:
            resultado.append(izquierda[i])
            i += 1

        else:
            resultado.append(derecha[j])
            j += 1

    resultado.extend(izquierda[i:])
    resultado.extend(derecha[j:])

    return resultado
```

```python
magnitudes = df["Magnitude"].tolist()

magnitudes_ordenadas = merge_sort(magnitudes)

print("Primeras 10 magnitudes ordenadas:")
print(magnitudes_ordenadas[:10])

print("\nÚltimas 10 magnitudes ordenadas:")
print(magnitudes_ordenadas[-10:])
```

```python
# Generar archivo de resultados de eficiencia

resultados_eficiencia = pd.DataFrame({
    "metodo": ["bucle_iterrows", "vectorizado_pandas"],
    "tiempo_segundos": [tiempo_bucle, tiempo_vectorizado],
    "magnitud_minima": [6.0, 6.0]
})

resultados_eficiencia["mejora_relativa"] = tiempo_bucle / tiempo_vectorizado

ruta_resultados = "resultados/metricas_eficiencia.csv"

resultados_eficiencia.to_csv(ruta_resultados, index=False)

print("Archivo generado correctamente:")
print(ruta_resultados)

resultados_eficiencia
```
