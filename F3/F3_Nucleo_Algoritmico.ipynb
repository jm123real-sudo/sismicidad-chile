## Medición de eficiencia algorítmica

Se comparan dos implementaciones para identificar sismos con magnitud mayor o igual a 6.0:

- Implementación con recorrido fila por fila.
- Implementación vectorizada con pandas.

El objetivo es medir el tiempo de ejecución y justificar la elección del método más eficiente.

```python
def buscar_sismos_bucle(df, magnitud_minima):
    """
    Busca sismos fuertes recorriendo el DataFrame fila por fila.
    Esta implementación es menos eficiente porque evalúa cada registro individualmente en Python.
    """
    resultados = []

    for _, fila in df.iterrows():
        if fila["Magnitude"] >= magnitud_minima:
            resultados.append(fila)

    return resultados


def buscar_sismos_vectorizado(df, magnitud_minima):
    """
    Busca sismos fuertes usando operaciones vectorizadas de pandas.
    Esta implementación aprovecha optimizaciones internas y suele ser más eficiente.
    """
    return df[df["Magnitude"] >= magnitud_minima]
```

```python
tiempo_bucle = timeit.timeit(
    lambda: buscar_sismos_bucle(df, 6.0),
    number=10
)

tiempo_vectorizado = timeit.timeit(
    lambda: buscar_sismos_vectorizado(df, 6.0),
    number=10
)

print("Tiempo con bucle:", tiempo_bucle)
print("Tiempo vectorizado:", tiempo_vectorizado)
print("Diferencia:", tiempo_bucle / tiempo_vectorizado)
```

## Algoritmo recursivo: Merge Sort

Para demostrar la aplicación de recursividad se implementa el algoritmo Merge Sort sobre la variable Magnitude.

Este algoritmo utiliza la estrategia divide y vencerás, dividiendo el problema en subproblemas más pequeños hasta alcanzar un caso base y posteriormente combinando los resultados ordenados.

```python
def merge_sort(lista):

    if len(lista) <= 1:
        return lista

    medio = len(lista) // 2

    izquierda = merge_sort(lista[:medio])
    derecha = merge_sort(lista[medio:])

    return combinar(izquierda, derecha)


def combinar(izquierda, derecha):

    resultado = []
    i = 0
    j = 0

    while i < len(izquierda) and j < len(derecha):

        if izquierda[i] <= derecha[j]:
            resultado.append(izquierda[i])
            i += 1

        else:
            resultado.append(derecha[j])
            j += 1

    resultado.extend(izquierda[i:])
    resultado.extend(derecha[j:])

    return resultado
```

```python
magnitudes = df["Magnitude"].tolist()

magnitudes_ordenadas = merge_sort(magnitudes)

print("Primeras 10 magnitudes ordenadas:")
print(magnitudes_ordenadas[:10])

print("\nÚltimas 10 magnitudes ordenadas:")
print(magnitudes_ordenadas[-10:])
```

# Programación Orientada a Objetos

Con el objetivo de mejorar la modularidad, reutilización y mantenibilidad del código, se implementa una clase encargada de gestionar las operaciones principales sobre el dataset sísmico.

Esta aproximación permite encapsular los datos y los métodos asociados, facilitando futuras extensiones del sistema.

```python
class AnalizadorSismico:

    def __init__(self, dataframe):
        self.df = dataframe

    def cantidad_registros(self):
        return len(self.df)

    def magnitud_maxima(self):
        return self.df["Magnitude"].max()

    def profundidad_promedio(self):
        return self.df["Depth"].mean()

    def contar_sismos_fuertes(self, magnitud_minima):
        return len(
            self.df[self.df["Magnitude"] >= magnitud_minima]
        )
```

```python
analizador = AnalizadorSismico(df)

print("Cantidad de registros:")
print(analizador.cantidad_registros())

print("\nMagnitud máxima:")
print(analizador.magnitud_maxima())

print("\nProfundidad promedio:")
print(round(analizador.profundidad_promedio(), 2))

print("\nCantidad de sismos >= 6.0:")
print(analizador.contar_sismos_fuertes(6.0))
```

## Herencia y polimorfismo

Para fortalecer el diseño orientado a objetos, se implementa una clase avanzada que hereda de la clase base `AnalizadorSismico`.

La clase hija reutiliza métodos existentes y modifica el comportamiento del método `resumen()`, demostrando herencia y polimorfismo.

```python
class AnalizadorSismicoBase:

    def __init__(self, dataframe):
        self.df = dataframe

    def resumen(self):
        return {
            "registros": len(self.df),
            "magnitud_maxima": self.df["Magnitude"].max()
        }


class AnalizadorSismicoAvanzado(AnalizadorSismicoBase):

    def resumen(self):
        resumen_base = super().resumen()

        resumen_base["profundidad_promedio"] = round(self.df["Depth"].mean(), 2)
        resumen_base["sismos_fuertes"] = len(self.df[self.df["Magnitude"] >= 6.0])
        resumen_base["magnitud_promedio"] = round(self.df["Magnitude"].mean(), 2)

        return resumen_base
```

```python
analizador_base = AnalizadorSismicoBase(df)
analizador_avanzado = AnalizadorSismicoAvanzado(df)

print("Resumen clase base:")
print(analizador_base.resumen())

print("\nResumen clase avanzada:")
print(analizador_avanzado.resumen())
```
